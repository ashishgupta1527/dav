```python
import pandas as pd
import matplotlib.pyplot as mlp
import numpy as np
import seaborn as sn
```

```python
df1 = pd.read_csv("bhp.csv")
```

```python
df = df1.head(10)
df
```

```python
sn.set(rc={'figure.figsize':(15,7.5)})
sn.histplot(x='location', data=df)
```

```python
sn.set(rc={'figure.figsize':(15,7.5)})
sn.boxplot(x='price', data=df)
```

```python
x= df[['location']]
electronic = df.loc[df['location'] == 'Electronic City Phase II'].count()[0]
chikka = df.loc[df['location'] == 'Chikka Tirupathi'].count()[0]
uttarahalli = df.loc[df['location'] == 'Uttarahalli'].count()[0]
lingadheeranahalli = df.loc[df['location'] == 'Lingadheeranahalli'].count()[0]
kothanur = df.loc[df['location'] == 'Kothanur'].count()[0]
whitefield = df.loc[df['location'] == 'Whitefield'].count()[0]
old = df.loc[df['location'] == 'Old Airport Road'].count()[0]
rajaji = df.loc[df['location'] == 'Rajaji Nagar'].count()[0]
marathahalli = df.loc[df['location'] == 'Marathahalli'].count()[0]
gandhi = df.loc[df['location'] == 'Gandhi Bazar'].count()[0]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
labels = ['Electronic City Phase II', 'Chikka Tirupathi', 'Uttarahalli', 'Lingadheeranahalli', 'Kothanur', 'Whitefield', 'Old Airport Road', 'Rajaji Nagar', 'Marathahalli', 'Gandhi Bazar']
plt.pie([electronic, chikka, uttarahalli, lingadheeranahalli, kothanur, whitefield, old, rajaji, marathahalli, gandhi], labels = labels, autopct='%.1f %%')

plt.title('Plot in Region')
plt.show
```
